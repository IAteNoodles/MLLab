```python
import numpy as np, matplotlib.pyplot as plt
from collections import Counter

d = np.random.rand(100)
tr, te = d[:50], d[50:]
lab = ['Class1' if x<=0.5 else 'Class2' for x in tr]

def knn(p,k):
    return Counter(y for _,y in sorted((abs(p-x),y) for x,y in zip(tr,lab))[:k]).most_common(1)[0][0]

for k in [1,2,3,4,5,20,30]:
    pred = [knn(x,k) for x in te]

    print(f"\nk={k}")
    for i,(x,y) in enumerate(zip(te,pred),51):
        print(f"x{i}: {x:.4f} -> {y}")

    c1 = [x for x,y in zip(te,pred) if y=='Class1']
    c2 = [x for x,y in zip(te,pred) if y=='Class2']

    plt.figure(figsize=(10,6))
    plt.scatter(tr,[0]*50,c=['blue' if y=='Class1' else 'red' for y in lab])
    plt.scatter(c1,[1]*len(c1),c='blue',marker='x')
    plt.scatter(c2,[1]*len(c2),c='red',marker='x')

    plt.title(f"k-NN (k={k})")
    plt.xlabel("Data Points")
    plt.ylabel("Classification Level")
    plt.grid()
    plt.show()
```
