```
import pandas as pd

df = pd.read_csv("Training_Data.csv")
X, y = df.iloc[:,:-1].values, df.iloc[:,-1].values

S = X[0].copy()
G = [['?'] * len(S)]

for x, label in zip(X, y):
    if label.lower() == 'yes':
        S = [s if s == v else '?' for s, v in zip(S, x)]
        G = [g for g in G if all(a == '?' or a == b for a, b in zip(g, x))]
    else:
        G += [[s if s != x[i] else '?' for i, s in enumerate(S)]]

print("Specific Hypothesis:", S)
print("General Hypothesis:")
for g in G: print(g)
```
